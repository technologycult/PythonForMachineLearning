```python
import pandas as pd
```

# creating and initializing a list

```python
values= [[60,78,65,52,85],
         [75,85,62,61,60],
         [89,76,56,51,35],
         [40,60,70,40,45],
         [50,70,75,46,85],
         [45,66,77,49,78]]
```

```python
df1 = pd.DataFrame(values,columns=[
                                  'English',
                                  'Physics',
                                  'Chemistry',
                                  'Maths',
                                  'Biology'],
index=['Majid','Mohsin','Suresh','Suman','Monish','Vaani'])
```

```python
df1
```

# Task No 1
# As because the maths paper was very tough
# it was decided that 10 grace marks will be provided.
# Applying Lambda function to a single column

```python
df1['Maths'] = df1['Maths'].apply(lambda x : x + 10)
df1
```

# Task No 2
# Applying Lambda function to Multiple column

```python
df1 = df1.apply(lambda x : x + 5)
df1
```

# Task No 3
# Applying Lambda function to a Multiple column and get a new column call Total Marks

```python
df1['Total'] = df1.apply(lambda x : x['English'] +
                                  x['Physics'] +
                                  x['Chemistry'] +
                                  x['Maths'] +
                                  x['Biology'], axis=1)
df1
```

# Task No 4
# Applying Lambda function to a Multiple column and get a new column call Total Percentage

```python
df1['Percentage'] = df1.apply(lambda x : (x['English'] +
                                  x['Physics'] +
                                  x['Chemistry'] +
                                  x['Maths'] +
                                  x['Biology'])/5, axis=1)
df1
```

```python
df1.drop(['Total','Percentage'],inplace=True,axis=1)
df1
```

# Task No 5
# Applying Lambda function to a Single row

```python
df1 = df1.apply(lambda x: x -10 if x.name in ['Suresh'] else x, axis=1 )
df1
```

# Task no 6
# Applying Lambda function to a Single column

```python
df1 = df1.apply(lambda x: x-6 if x.name in ['Chemistry'] else x, axis=0)
df1
```

# Task No 7
# Applying Lambda function to a Single column based on conditions

```python
df1['Physics'] = df1['Physics'].apply(lambda x: x -10 if x>=75 else x)
df1
```

# Task No 8
# How to update Specific Cell of a DataFrame

```python
df1['Biology'] = df1['Biology'].apply(lambda x : x + 10 if x<=39 else x)
df1
```

# Task No 9
# Applying Lambda function to multiple rows based on index

```python
df1 = df1.apply(lambda x : x + 5 if x.name in ['Suman','Vaani'] else x, axis=1)
df1
```

# Task No 10
# Applying Lambda function to multiple columns

```python
df1 = df1.apply(lambda x : x + 3 if x.name in ['Biology','Maths'] else x, axis=0)
df1
```

# Task No 11
# Create new column Pass/Fail

```python
df1['Grade_english'] = df1['English'].apply(lambda x : 'Pass' if x > 50 else 'Fail' )
df1
```

```python
df1.drop(['Grade_english'],inplace=True,axis=1)
df1
```

# Task No 12
# Create a new column that has the grade instead of marks
# A -- marks greater than equal to 80
# B -- marks greater than equal to 60 and less than 80
# C -- marks greater than equal to 40 and less than 60
# F -- marks less than 40

```python
df1['Grade_english'] = df1['English'].apply(lambda x : 'A' if x>= 80 else
                                                       ('B' if x>= 60 else
                                                       ('C' if x>= 40 else 'F')))
df1
```

```python
df1.drop(['Grade_english'],inplace=True,axis=1)
df1
```

# Task no 13
# Get rows with maximum marks for each subject and add it in the existing dataframe

```python
max1 = df1.apply(lambda x: max(x))
df1.loc['Max'] = max1
df1
```

# Task no 14
# Get rows with minimum marks for each subject and add it in the existing dataframe

```python
min1 = df1.apply(lambda x: min(x))
df1.loc['Min'] = min1
df1
```
